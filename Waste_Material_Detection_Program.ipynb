```python
from google.colab import files
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
import io
import PIL
model = load_model('/content/drive/MyDrive/Dataset/Waste_Model.h5')
def predict_image_category(img_path):
    img = image.load_img(img_path, target_size=(150,150))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0
    prediction = model.predict(img_array)
    return prediction
def handle_upload(input_file):
    image_bytes = io.BytesIO(input_file)
    img = PIL.Image.open(image_bytes)
    display(img)
    prediction = predict_image_category(image_bytes)
    categories = ['cardboard', 'glass', 'metal', 'paper', 'plastic']
    predicted_category = categories[np.argmax(prediction)]
    print('Predicted Category:', predicted_category)
uploaded_files = files.upload()
for uploaded_file in uploaded_files.keys():
    handle_upload(uploaded_files[uploaded_file])
```
